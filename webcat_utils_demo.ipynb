```python
import numpy as np
from webcat_utils import VidViewer
```

#### Instantiate VidViewer class with a url

```python
url = "http://webcat-video.axds.co/buxtoncoastalcam/raw/2019/2019_11/2019_11_13/buxtoncoastalcam.2019-11-13_1000.mp4"
vr = VidViewer(url)
```

#### View properties of video

```python
print(f"Frame width: {vr.width}")
print(f"Frame height: {vr.height}")
print(f"Total frames: {vr.frames}")
print(f"FPS: {vr.fps}")
```

#### View some frames

```python
vr.plot_frames(frames=[0, 100, vr.frames - 1])
```

```python
# view average frame
step = 500 # average of every N frames
vr.plot_average_frame(step)
```

#### Download video to local file

```python
vr.download_url(fout='../data/videos/buxtoncoastalcam.2019-11-13_1000.mp4')
```

#### Save frames of video to .jpg files

```python
vr.save_frames(delta_t=600, fout="./")
```

### Looping

The `VidViewer()` class is constructed such that you can pass new urls without having to re-instantiate the class (could be good for looping over several vide0s in a list of urls)

```python
url_1 = "http://webcat-video.axds.co/buxtoncoastalcam/raw/2019/2019_11/2019_11_13/buxtoncoastalcam.2019-11-13_1000.mp4"
url_2 = "http://webcat-video.axds.co/buxtoncoastalcam/raw/2019/2019_11/2019_11_13/buxtoncoastalcam.2019-11-13_2000.mp4"
```

```python
vr = VidViewer(url_1)
vr.plot_frames()
```

```python
vr.fin = url_2
vr.plot_frames()
```
