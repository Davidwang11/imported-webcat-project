```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import buoypy as bp # https://github.com/nickc1/buoypy
%matplotlib inline
pd.set_option("display.max_rows", None)
```

```python
buoy = 41025 #DIAMOND SHOALS
year = 2019

H = bp.historic_data(buoy,year)
df = H.get_stand_meteo()
SSdf= df[df.WVHT != 99.00]

# plotting
Bplot =SSdf.WVHT.plot()
Bplot.set_ylabel('Wave Height (m)',fontsize=14)
```

```python
#early 2019 data that is missing from Diamond Shoals
buoy = 44095 #Oregon Inlet
year = 2019

H = bp.historic_data(buoy,year)
df = H.get_stand_meteo()
SSdf= df[df.WVHT != 99.00]

# plotting All
Bplot =SSdf.WVHT.plot()
Bplot.set_ylabel('Wave Height (m)',fontsize=14)
```

```python
# plotting just the first 6 mos of 2019
Bplot =SSdf.WVHT.plot()
Bplot.set_ylabel('Wave Height (m)',fontsize=14)
Bplot.set_xlim("2019-01-01 00:00:00", "2019-05-01 00:00:00")
```

```python
#realtime, just the last 45 days (to look at May 1 2020 event)
buoy = 41025 #DIAMOND SHOALS
DS = bp.realtime(buoy) #last 45 days

DS = DS.txt()

# plotting
Bplot =DS.WVHT.plot()
Bplot.set_ylabel('Wave Height (m)',fontsize=14)
```
